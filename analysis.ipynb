```python
languages = pd.DataFrame(
    {"language": courses["language"].unique(), "value": 10})

for index, row in languages.iterrows():
    languages.loc[languages['language'] == row["language"], ['value']]= len(courses[courses['language'] == row["language"]])

ex = languages[languages['value'] < 3000]
countex=ex['value'].sum()
languages = languages[languages['value'] >= 3000]

new_row = {'language':'Other', 'value':countex}
languages=languages.append(new_row, ignore_index=True)

```

```python
pal_ = list(sns.color_palette(palette='pastel', n_colors=len(languages)).as_hex())
                              
plot = languages.plot.pie(y='value', labels=languages['language'], colors=pal_, autopct='%1.1f%%', pctdistance=0.8)
plt.legend(bbox_to_anchor=(1, 1), loc=2, frameon=False)

plt.savefig('plot\languages.png', bbox_inches='tight')
```

```python
price = pd.DataFrame({"price": ['Free', '<10$', '<20$', '<50$', '50+$'], "value": 0})

price.loc[price['price'] == 'Free', ['value']]= len(courses[courses['price'] ==0])
price.loc[price['price'] == '<10$', ['value']]=  len(courses[courses['price'].between(1, 10)])
price.loc[price['price'] == '<20$', ['value']]=  len(courses[courses['price'].between(10, 20)])
price.loc[price['price'] == '<50$', ['value']]=  len(courses[courses['price'].between(20, 50)])
price.loc[price['price'] == '50+$', ['value']]=  len(courses[courses['price'] >50])
```

```python
pal_ = list(sns.color_palette(palette='pastel', n_colors=len(price)).as_hex())
                              
plot = price.plot.pie(y='value', labels=price['price'], colors=pal_, autopct='%1.1f%%', pctdistance=0.8)
plt.legend(bbox_to_anchor=(1, 1), loc=2, frameon=False)

plt.savefig('plot\prices.png', bbox_inches='tight')
```
